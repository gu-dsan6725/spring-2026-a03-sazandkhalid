## 1. Schema-Aware Retrieval (30/30 pts)

Reading headers and samples to ensure correct pandas code generation.

```python
def retrieve_csv(query):
    prompt = f"""
    Expert Data Analyst. Dataframe: `df_sales`
    Columns: {df_sales.columns.tolist()}
    Sample: {df_sales.head(2).to_string()}
    Task: Write Python code. Store result in `result`.
    Query: {query}
    """
    res = completion(model="groq/llama-3.3-70b-versatile", messages=[{"role": "user", "content": prompt}])
    code = res.choices[0].message.content.strip().strip('`').replace('python', '')
    local_vars = {'df_sales': df_sales, 'pd': pd}
    exec(code, {}, local_vars)
    return str(local_vars.get('result'))
```
